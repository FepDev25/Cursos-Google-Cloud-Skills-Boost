```python
import tensorflow as tf
```

```python
# Tensores

# Escaleres 
a = tf.constant(5)
b = tf.constant(3)
print("Escalar a:", a)
```

```python
# Vectores
v = tf.constant([5, 3, 8, 2])
print("Vector v:", v)
```

```python
# Matrices
m = tf.constant([[5, 3, 8],
                 [2, 4, 1],
                 [7, 9, 6]])
print("Matriz m:\n", m)
```

```python
# 3D Tensores
t = tf.constant([[[5, 3, 8],
                  [2, 4, 1]],
                    [[7, 9, 6],
                    [0, 2, 4]]])
print("3D Tensor t:\n", t)
```
